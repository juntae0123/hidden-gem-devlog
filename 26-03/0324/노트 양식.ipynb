## 🚨 트러블슈팅 (문제 및 해결)
- **문제 1: 포트 5432 연결 실패**
  - **에러:** psycopg2.OperationalError: Connection refused
  - **원인:** 도커 엔진은 켜져 있으나 실제 DB 컨테이너가 Started 상태가 아니었음.
  - **해결:** 루트 폴더로 이동 후 docker-compose up -d 명령어로 서버 가동.
- **문제 2: "vector" 타입 미존재 에러**
  - **에러:** psycopg2.errors.UndefinedObject: type "vector" does not exist
  - **원인:** DB 서버는 떴지만 pgvector 확장(Extension) 기능을 활성화하지 않아 VECTOR 타입을 인식 못 함.
  - **해결:** init_db.py에서 CREATE EXTENSION IF NOT EXISTS vector를 테이블 생성 전에 실행하도록 로직 수정.
