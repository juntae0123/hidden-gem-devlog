# 583개의 게임 데이터를 모두 384차원의 밀집 벡터로 변환한 뒤, 그 벡터(배열) 데이터를 깨짐 없이 보존

```python
import pandas as pd
import os
import sys
from sentence_transformers import SentenceTransformer
from dotenv import find_dotenv

# [1. 프로젝트 루트 경로 동적 탐색]
env_path = find_dotenv()
project_root = os.path.dirname(env_path)

input_path = os.path.join(project_root, "data", "hidden_gem_analyzed_final.csv")
# 💡 핵심: 벡터(배열) 데이터를 원형 그대로 보존하기 위해 .pkl 포맷 사용
output_path = os.path.join(project_root, "data", "hidden_gem_embedded.pkl") 

# [2. 분석 완료된 데이터 로드]
if not os.path.exists(input_path):
    print(f"❌ 파일을 찾을 수 없습니다: {input_path}")
    sys.exit()

df = pd.read_csv(input_path)
print(f"✅ 데이터 로드 완료: 총 {len(df)}건")

# [3. 다국어 밀집 벡터(Dense Vector) 모델 로드]
print("🧠 다국어 AI 임베딩 모델 로딩 중...")
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

# [4. 텍스트 병합 (Data Preprocessing)]
# 왜 합치나요?: 모델이 게임의 전체적인 문맥을 384차원에 풍부하게 담아내게 하기 위함입니다.
df['combined_text'] = df['name'].fillna("") + " : " + df['genres'].fillna("") + " : " + df['description'].fillna("")

# [5. 임베딩 변환 (Encoding)]
print(f"🔄 총 {len(df)}개 게임의 의미를 다차원 벡터로 압축합니다... (수 분 소요)")
texts_to_encode = df['combined_text'].tolist()

# show_progress_bar=True 로 변환 진행 상황을 눈으로 확인합니다.
embeddings = model.encode(texts_to_encode, show_progress_bar=True)

# [6. 데이터프레임에 벡터값(좌표) 저장]
# 모델이 뱉어낸 고차원 배열을 리스트로 변환하여 새로운 컬럼에 넣습니다.
df['embedding'] = embeddings.tolist()

# [7. Pickle 파일로 영구 저장]
df.to_pickle(output_path)
print(f"\n🎉 완벽합니다! 밀집 벡터 변환이 완료되어 안전하게 저장되었습니다: {output_path}")
```
