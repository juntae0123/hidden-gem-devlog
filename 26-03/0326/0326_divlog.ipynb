## 🛠 진행 상황 및 핵심 기록
- **백엔드 점수 공식 리팩토링:** `(metadata * 0.8) + (similarity * 0.4)` 공식을 통해 100점 돌파(`LEGENDARY`) 등급이 자연스럽게 발생하도록 조정.
- **Exact Match 프리패스:** `is_exact_match`인 경우 장르 일치 여부와 상관없이 무조건 메인 탭으로 분류하고 +50점 가산점 부여.
- **0등석 HeroBanner:** `page.tsx`에서 데이터를 분리하여 최상단에 배치. 아코디언(접기/펴기) 없이 설명과 레이더 차트가 즉시 보이는 가로형 레이아웃 적용.
- **등급별 비주얼 아이덴티티:**
    - `Exact Match`: 영롱한 **금색(Gold)** 네온 글로우 + 👑 왕관 뱃지
    - `LEGENDARY (100+)`: 강렬한 **빨간색(Red)** 네온 글로우
    - `EPIC`: 고급스러운 **보라색(Purple)** 네온 글로우

## 🚨 트러블슈팅 (문제 및 해결)
- **문제 1: '림월드' 검색 시 엉뚱한 인기 게임(위쳐3 등)이 메인 탭 차지 및 림월드 유배 사태**
    - **원인:** DB의 한글 장르("전략")와 GPT가 추출한 영어 장르("Strategy")가 텍스트 불일치로 판정되어 진짜 결과들이 서브 탭으로 밀려남. 비어있는 메인 탭을 유사도 50%의 Fallback(인기 게임) 데이터가 채움.
    - **해결:** `main.py` 내 `check_genre_match` 함수에서 한글 키워드 직접 비교 로직을 추가하고, `is_exact_match`인 경우 장르와 무관하게 메인 탭 입성을 보장함.

- **문제 2: Tailwind CSS v4.0 도입에 따른 문법 및 경로 에러**
    - **현상:** `@tailwind base` 지시어 미작동 및 `@/types` 절대 경로 인식 불가.
    - **원인:** Tailwind v4의 사양 변경 및 Next.js 프로젝트 설정 상의 경로 매핑 문제.
    - **해결:** `globals.css`를 최신 문법인 `@import "tailwindcss";`로 수정하고, 모든 임포트 경로를 상대 경로(`./`, `../`)로 전면 교체.

- **문제 3: HeroBanner 컴포넌트 Import 불가 (대소문자 캐시 지옥)**
    - **현상:** 파일명을 고쳐도 `Cannot find module './components/herobanner'` 에러 지속.
    - **원인:** 파일명은 `herobanner.tsx`인데 Import는 `HeroBanner`로 함. Next.js 빌드 캐시(`.next`)가 이전 파일명을 기억하고 있어 대소문자 수정을 인식 못 함.
    - **해결:** 파일명을 `HeroBanner.tsx`로 확실히 대문자화하고, `.next` 캐시 폴더를 삭제한 후 서버를 재시작하여 해결.
